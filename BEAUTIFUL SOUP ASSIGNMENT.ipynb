```python
#Please visit  https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded
# articles/  and scrap- 
# a) Paper title 
# b)    date 
# c) Author

import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content
    soup = BeautifulSoup(response.content, 'html.parser')

    article_containers = soup.find_all('div', class_='panel') # Adjust class name here
    print("Total articles:",len(article_containers))
    
    for container in article_containers:
        # Extract title (adjust the tag and class based on your inspection)
        title = container.find('h2', class_='h5 article-title').get_text(strip=True) # Adjust tag and class
        
        # Extract date (adjust the tag and class based on your inspection)
        date = container.find('p', class_="article-date").text # Adjust tag and class
        
        # Extract authors (adjust the tag and class based on your inspection)
        authors = container.find('p', class_='article-authors').text # Adjust tag and class
        
        print(f'Title: {title}')
        print(f'Date: {date}')
        print(f'Authors: {authors}')
        print('-' * 40)
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
    
    
    
    
```

```python
#4) Write a python program to scrape first 10 product details which include product name , price , Image URL from 
#https://www.bewakoof.com/bestseller?sort=popular ...

import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://www.bewakoof.com/bestseller?sort=popular'

# Send an HTTP request to the URL
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
product_containers = soup.find_all('div', class_='plp-product-card',limit=10)
  
for container in product_containers:
        # Extract the product name
        product_name = container.find('div', class_='productNaming bkf-ellipsis').get_text(strip=True)
          # Extract the product price
        product_price = container.find('div', class_='discountedPriceText clr-p-black false').get_text(strip=True)
        
        # Extract the image URL
        image_tag = container.find('img', class_='productImgTag')
        image_url = image_tag['src']
        
        print(f'Product Name: {product_name}')
        print(f'Discounted Price: {product_price}')
        print(f'Image URL: {image_url}')
        print('-' * 40)

```

```python
# Please visit https://www.cnbc.com/world/?region=world and scrap- 
# a)   headings 
# b)  date 
# c) News link 
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://www.cnbc.com/world/?region=world'

# Send an HTTP request to the URL
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

product_containers = soup.find_all('li',class_='LatestNews-item')
print ("Latest ",len(product_containers)," News")  

for container in product_containers:
        # Extract the product name
        heading = container.find('a', class_='LatestNews-headline').get_text(strip=True)
          # Extract the product price
        date = container.find('time', class_='LatestNews-timestamp').get_text(strip=True)
        
        # Extract the image URL
        news_tag = container.find('a', class_='LatestNews-headline')
        news_url=news_tag['href']
        
        print(f'Heading: {heading}')
        print(f'Date: {date}')
        print(f'News URL: {news_url}')
        print('-' * 40)
```
